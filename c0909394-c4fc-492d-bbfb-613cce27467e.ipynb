````python
def comma_to_dot_cow_buy(row):
    row['–¢–µ–∫—É—â–∞—è_–∂–∏—Ä–Ω–æ—Å—Ç—å,%'] = row['–¢–µ–∫—É—â–∞—è_–∂–∏—Ä–Ω–æ—Å—Ç—å,%'].replace(',','.')
    row['–¢–µ–∫—É—â–∏–π_—É—Ä–æ–≤–µ–Ω—å_–±–µ–ª–æ–∫,%'] = row['–¢–µ–∫—É—â–∏–π_—É—Ä–æ–≤–µ–Ω—å_–±–µ–ª–æ–∫,%'].replace(',','.')
    return row
cow_buy = cow_buy.apply(comma_to_dot_cow_buy, axis=1)
````

````python
def comma_to_dot_ferma_main(row):
    row['–≠–ö–ï (–≠–Ω–µ—Ä–≥–µ—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–º–æ–≤–∞—è –µ–¥–∏–Ω–∏—Ü–∞)'] = row['–≠–ö–ï (–≠–Ω–µ—Ä–≥–µ—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–º–æ–≤–∞—è –µ–¥–∏–Ω–∏—Ü–∞)'].replace(',','.')
    row['–°–ü–û (–°–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)'] = row['–°–ü–û (–°–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)'].replace(',','.')
    row['–ñ–∏—Ä–Ω–æ—Å—Ç—å,%'] = row['–ñ–∏—Ä–Ω–æ—Å—Ç—å,%'].replace(',','.')
    row['–ë–µ–ª–æ–∫,%'] = row['–ë–µ–ª–æ–∫,%'].replace(',', '.')
    return row

ferma_main = ferma_main.apply(comma_to_dot_ferma_main, axis=1)
````

````python
ferma_main['–≠–ö–ï (–≠–Ω–µ—Ä–≥–µ—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–º–æ–≤–∞—è –µ–¥–∏–Ω–∏—Ü–∞)'] = ferma_main['–≠–ö–ï (–≠–Ω–µ—Ä–≥–µ—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–º–æ–≤–∞—è –µ–¥–∏–Ω–∏—Ü–∞)'].astype(float)
ferma_main['–°–ü–û (–°–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)'] = ferma_main['–°–ü–û (–°–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)'].astype(float)
ferma_main['–ñ–∏—Ä–Ω–æ—Å—Ç—å,%'] = ferma_main['–ñ–∏—Ä–Ω–æ—Å—Ç—å,%'].astype(float)
ferma_main['–ë–µ–ª–æ–∫,%'] = ferma_main['–ë–µ–ª–æ–∫,%'].astype(float) 
cow_buy['–¢–µ–∫—É—â–∞—è_–∂–∏—Ä–Ω–æ—Å—Ç—å,%'] = cow_buy['–¢–µ–∫—É—â–∞—è_–∂–∏—Ä–Ω–æ—Å—Ç—å,%'].astype(float)
cow_buy['–¢–µ–∫—É—â–∏–π_—É—Ä–æ–≤–µ–Ω—å_–±–µ–ª–æ–∫,%'] = cow_buy['–¢–µ–∫—É—â–∏–π_—É—Ä–æ–≤–µ–Ω—å_–±–µ–ª–æ–∫,%'].astype(float)
````

## –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏

–î–∞–≤–∞–π—Ç–µ –æ—Ç–±–µ—Ä–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–¥—Ö–æ–¥—è—Ç –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, —ç—Ç–æ –ø—Ä–∏–∑–Ω–∞–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –∏–º–µ—é—Ç —Å–∏–ª—å–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å —Å —Ü–µ–ª–µ–≤—ã–º, –∞ —Ç–∞–∫–∂–µ –æ–Ω–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ö–æ—Ç—å —á—É—Ç—å-—á—É—Ç—å –ª–∏–Ω–µ–π–Ω–æ–π, —Ç–æ–≥–¥–∞ –Ω–∞–º –ø–æ–¥—Ö–æ–¥—è—Ç —Ç–∞–∫–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏:

`–ü–æ—Ä–æ–¥–∞`

–û—Å—Ç–∞–ª—å–Ω—ã–µ –∂–µ –ø—Ä–∏–∑–Ω–∞–∫–∏, —Ç–∞–∫–∏–µ –∫–∞–∫ –ñ–∏—Ä–Ω–æ—Å—Ç—å –∏ –°–æ–¥–µ—Ä–∂–∞–Ω–∏–µ –±–µ–ª–∫–∞ –Ω–µ –∏–º–µ–µ—Ç –ø—Ä–∏—á–∏–Ω–æ —Å–ª–µ–¥—Å—Ç–≤–µ–Ω–Ω—É—é –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å

````python
RANDOM_STATE = 42
X = ferma_main.drop(['id','–ñ–∏—Ä–Ω–æ—Å—Ç—å,%', '–ë–µ–ª–æ–∫,%','–£–¥–æ–π, –∫–≥'], axis=1)
y = ferma_main['–£–¥–æ–π, –∫–≥']
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state = RANDOM_STATE
)

cat_col_names = ['–ü–æ—Ä–æ–¥–∞', '–¢–∏–ø –ø–∞—Å—Ç–±–∏—â–∞','–ø–æ—Ä–æ–¥–∞ –ø–∞–ø—ã_–±—ã–∫–∞','–í–æ–∑—Ä–∞—Å—Ç']
num_col_names = ['–≠–ö–ï (–≠–Ω–µ—Ä–≥–µ—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–º–æ–≤–∞—è –µ–¥–∏–Ω–∏—Ü–∞)', '–°—ã—Ä–æ–π –ø—Ä–æ—Ç–µ–∏–Ω, –≥','–°–ü–û (–°–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)']

onehotencoder = OneHotEncoder(drop='first', sparse=False)

X_train_ohe = onehotencoder.fit_transform(X_train[cat_col_names])

X_test_ohe = onehotencoder.transform(X_test[cat_col_names])

encoder_col_names = onehotencoder.get_feature_names()                                     

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train[num_col_names])

X_test_scaled = scaler.transform(X_test[num_col_names])


X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)

X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)


X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)

X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)


X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)

X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)



first_model_lr = LinearRegression()

first_model_lr.fit(X_train, y_train)

predictions = first_model_lr.predict(X_test)

MSE = mean_squared_error(y_test, predictions)
MAE = mean_absolute_error(y_test, predictions)
RMSE = mean_squared_error(y_test, predictions, squared=False)

r2 = r2_score(y_test, predictions)
print("R2 LinearRegression =", r2)
print("MSE LinearRegression =", MSE)
print("MAE LinearRegression =", MAE)
print("RMSE LinearRegression =", RMSE)


residuals = y_test - predictions

y_pred_train = first_model_lr.predict(X_train)
errors = y_train - y_pred_train

alpha = 0.05
lower_quantile = errors.quantile(alpha)
upper_quantile = errors.quantile(1 - alpha / 2)

print(f"–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –ø—Ä–æ–≥–Ω–æ–∑–∞ (alpha={alpha}): [{lower_quantile:.2f}, {upper_quantile:.2f}]")


residuals.plot(kind='hist', bins=50, figsize=(12,5), grid=True)
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤ –º–æ–¥–µ–ª–∏')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞');
````

````python
RANDOM_STATE = 42
X = ferma_main.drop(['id','–£–¥–æ–π, –∫–≥','–í–∫—É—Å –º–æ–ª–æ–∫–∞','–≠–ö–ï (–≠–Ω–µ—Ä–≥–µ—Ç–∏—á–µ—Å–∫–∞—è –∫–æ—Ä–º–æ–≤–∞—è –µ–¥–∏–Ω–∏—Ü–∞)', '–°–ü–û (–°–∞—Ö–∞—Ä–æ-–ø—Ä–æ—Ç–µ–∏–Ω–æ–≤–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ)'], axis=1)
y = ferma_main['–í–∫—É—Å –º–æ–ª–æ–∫–∞']

X_train_5, X_test_5, y_train_5, y_test_5 = train_test_split(
    X,
    y,
    random_state=RANDOM_STATE
)
num_col_names_5 = ['–≠–ö–ï_square','–ë–µ–ª–æ–∫,%','–ñ–∏—Ä–Ω–æ—Å—Ç—å,%','–°—ã—Ä–æ–π –ø—Ä–æ—Ç–µ–∏–Ω, –≥']
cat_col_names_5 = ['–ü–æ—Ä–æ–¥–∞','–¢–∏–ø –ø–∞—Å—Ç–±–∏—â–∞','–ø–æ—Ä–æ–¥–∞ –ø–∞–ø—ã_–±—ã–∫–∞','–í–æ–∑—Ä–∞—Å—Ç','–°–ü–û_category']

encoder_4 = OneHotEncoder(sparse=False)
scaler_4 = StandardScaler()

X_train_scaled_5 = scaler_4.fit_transform(X_train_5[num_col_names_5])
X_test_scaled_5 = scaler_4.transform(X_test_5[num_col_names_5])

X_train_ohe_5 = encoder_4.fit_transform(X_train_5[cat_col_names_5])
X_test_ohe_5 = encoder_4.transform(X_test_5[cat_col_names_5])
encoder_names_5 = encoder_4.get_feature_names()

X_train_scaled_5 = pd.DataFrame(X_train_scaled_5, columns=num_col_names_5)
X_test_scaled_5 = pd.DataFrame(X_test_scaled_5, columns=num_col_names_5)

X_train_ohe_5 = pd.DataFrame(X_train_ohe_5, columns=encoder_names_5)
X_test_ohe_5 = pd.DataFrame(X_test_ohe_5, columns=encoder_names_5)

X_train_5 = pd.concat([X_train_scaled_5,X_train_ohe_5], axis=1)
X_test_5 = pd.concat([X_test_scaled_5, X_test_ohe_5], axis=1)

clf = LogisticRegression()
clf.fit(X_train_5, y_train_5)

y_pred_5 = clf.predict(X_test_5)
y_proba = clf.predict_proba(X_test_5)

accuracy = accuracy_score(y_test_5, y_pred_5)
precision = precision_score(y_test_5, y_pred_5)
recall = recall_score(y_test_5, y_pred_5)
print(f"Accuracy score {accuracy}")
print(f"Precision score {precision}")
print(f"Recall score {recall}")


probabilities_one_valid = clf.predict_proba(X_test_5)[:,1]

for i in np.arange(0.5, 0.9, 0.02).round(2):
    y_pred_6 = probabilities_one_valid > i
    if precision_score(y_test_5, y_pred_6) == 1:
        print('–ø–æ—Ä–æ–≥: ', i, 'precision: ', precision_score(y_test_5, y_pred_6))
        break



    
threshold = 0.8

y_pred_proba = clf.predict_proba(X_test_5)[:, 1]
y_pred_5_new = (y_pred_proba >= threshold).astype(int)

new_recall = recall_score(y_test_5, y_pred_5_new)
new_precision = precision_score(y_test_5, y_pred_5_new)
print(f"New Precision score {new_precision}")
print(f"New Recall score {new_recall}")





cm = confusion_matrix(y_test_5, y_pred_5_new)
plt.figure(figsize=(12,5))
sns.heatmap(cm, annot=True,fmt='d', cmap='Blues_r')
plt.ylabel('True Label')
plt.xlabel('Predicted');
````
